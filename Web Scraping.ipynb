```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_cities_by_GDP')
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
table = soup.find('table')
```

```python
table_head = table.find_all('th')
```

```python
table_head_text = [i.text.strip() for i in table_head]
```

```python
table_head_text
```

```python
table_row = table.find_all('tr')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_head_text)
```

```python
for row in table_row[1:]:
    rows = row.find_all('td')
    individual_row = [i.text.strip() for i in rows]

    length = len(df)
    df.loc[length] = individual_row
```

```python
df
```

```python
df.to_csv(r'C:\Users\User\cities_by_gdp.csv',index = False)
```
